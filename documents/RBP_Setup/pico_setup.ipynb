## Install Thonny

Thonny is the simplest option to run Python code on Raspberry Pi Pico. You can access Thonny's webpage and download by click [here](https://thonny.org). The latest version updated in this notebook was 4.0.2 (February 7th, 2023). Thonny has options to install in Windows, Linux, or Mac.

## Adding MicroPython

1. Open Thonny

2. Plug in the micros USB cable into the power port of RBP Pico. DO NOT CONNECT IT TO YOUR COMPUPTER YET.



2. Locate the BOOTSEL button on the RBP Pico. 

!

3. Press and hold it while you connect the other end of the USB cable to your computer. This action will configure the RBP Pico into USB mass storage device mode.

4. Locate the bottom right-hand corner of Thonny and look where it says "Local Python". Click it and select MicroPython (Raspberry Pi Pico).



5. A new window will popup with three prompts:
    * Target Volume: Select--> RP1- RP2 D
    * MicroPython Variant: Select--> Pico / Pico H
    * Version: 1.91 (this might change with time)
    * Click Install

6. Now, you will configure the Interpreter. Go to the "Run--> Configure Interpreter..
    * Locate the Interpreter Tab
    * Which kinda of interpreter...? Select: MicroPython (Raspberry Pi Pico)
    * Port Select: Try to detect port automatically.
    * Hit Ok



## Install Picozero

1. Inside Thonny: Go to Tools-->Manage Packages

2. Search: picozero

3. Click hyperlink: picozero

4. Install

5. Close

### You are all set!
